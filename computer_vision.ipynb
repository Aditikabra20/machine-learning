```
from google.colab.patches import cv2_imshow
```

```
# show image
cv2_imshow(img)
```

```
# 1 is show the actual color channel --grb or rjb
img1=cv2.imread('download.jpg',1)
cv2_imshow(img1)
```

```
# 0 no coloe channel --black and white
img2=cv2.imread('download.jpg',0)
cv2_imshow(img2)
```

```
# show the transparency 
img3=cv2.imread('download.jpg',-1)
cv2_imshow(img3)
```

```
# split image
x,y,z=cv2.split(img)
```

```
cv2_imshow(x)
```

```
cv2_imshow(y)
```

```
cv2_imshow(z)
```

```
# crop the image
cv2_imshow(img[100:200,100:200])
```

```
# create the line
a=cv2.line(img,(0,0),(200,200),(0,75,0),3)
```

```
cv2_imshow(a)
```

```
b=cv2.rectangle(img,(0,0),(50,50),(0,200,255),9)
```

```
# create the rectangle
cv2_imshow(b)
```

```
c=cv2.rectangle(img,(0,0),(50,50),(0,200,255),-1)
cv2_imshow(c)
```
