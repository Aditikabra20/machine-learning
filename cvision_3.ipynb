```
import cv2
```

```
from google.colab.patches import cv2_imshow
```

```
img1=cv2.imread('download1.jpg')
```

```
cv2_imshow(img1)
```

```
img2=cv2.imread('download2.jpg')
```

```
cv2_imshow(img2)
```

```
#absolute diff between two images
img3=cv2.absdiff(img2,img1)
```

```
cv2_imshow(img3)
```

```
# bitwise diff between two images
img4=cv2.bitwise_and(img1,img2)
```

```
cv2_imshow(img4)
```

```
import numpy as np
```

```
# add two images
# axis=0 add images vertically
img5=np.concatenate((img1,img2),axis=0)
```

```
cv2_imshow(img5)
```

```
# add two images 
# axis=1 add images horizontally
img6=np.concatenate((img1,img2),axis=1)
```

```
cv2_imshow(img6)
```
