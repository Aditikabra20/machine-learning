```
import pandas as pd
```

```
# loading data
df=pd.read_csv('social.csv')
```

```
#schema
df.info()
```

```
# top 5 rows
df.head(5)
```

```
# removing user id
fdata=df.iloc[:,2:].values
fdata
```

```
df.head(5)
```

```
# now we can apply training test
features=fdata[0:,0:2]
features
```

```
label=fdata[0:,2]
label
```

```
from sklearn.model_selection import train_test_split
```

```
X,x,Y,y=train_test_split(features,label,test_size=0.2)
```

```

# now apply feature scaling
from sklearn.preprocessing import StandardScaler
```

```
sc=StandardScaler()
```

```
# apply scaling in training data only
X=sc.fit_transform(X)
X
```

```
#testing data scaling
x=sc.transform(x)
x
```

```
#knn
from sklearn.neighbors import KNeighborsClassifier
```

```
#randon forest
from sklearn.ensemble import RandomForestClassifier
```

```
# call rf with number of decision tree
rclf=RandomForestClassifier(n_estimators=20)
```

```
# now  train
trained=rclf.fit(X,Y)
```

```
# now predicting
output=trained.predict(x)
output
```

```
y
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y,output)
```
