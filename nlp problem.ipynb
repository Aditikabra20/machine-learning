**Web data scrapping**

```
import requests
```

```
data=requests.get('https://en.wikipedia.org/wiki/Swami_Vivekananda')
```

```
data
```

```
textdata=data.text
```

```
from bs4 import BeautifulSoup
```

```
soup=BeautifulSoup(textdata,'html5lib')
```

```
soup.text
```

```
type(soup)
```

```
#actual data
clean_data=soup.get_text()
clean_data
```

```
scrape_data=clean_data.split()
scrape_data
```

**StopWords**

```

import nltk
from nltk.corpus import stopwords
```

```
nltk.download('stopwords')
```

```
useful_data=[i for i in scrape_data if i.lower() not in stopwords.words('english')]
```

```
import matplotlib.pyplot as plt
```

```
graph=nltk.FreqDist(useful_data)
```

```
graph.plot(20)
```


**Tokenization**

```
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```
nltk.download('punkt')
```

```
sent=sent_tokenize(clean_data)
```

```
len(sent)
```

```
word=word_tokenize(clean_data)
```

```
len(word)
```

```
word_graph=nltk.FreqDist(word)
```

```
word_graph.plot(20)
```

**stemming**

```
from nltk import stem
```

```
from nltk.stem import PorterStemmer
```

```
port=PorterStemmer()
```

```
for i in useful_data:
  print(i,"   stem word -->      ",port.stem(i))
  
```

**lemminiztion**

```
from nltk.stem import WordNetLemmatizer
```

```
nltk.download('wordnet')
```

```
wl=WordNetLemmatizer()
```

```
for i in useful_data:
  print(i,"         lemmatize word  -->    ",wl.lemmatize(i))
```
